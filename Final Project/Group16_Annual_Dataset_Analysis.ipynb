```python
# Convert all object columns to float64
energy_annual = energy_annual.astype({col: 'float64' for col in energy_annual.select_dtypes(include='object').columns})
energy_annual.dtypes
```

```python
energy_annual['Annual Total'] = pd.to_datetime(energy_annual['Annual Total'], format='%Y')
```

  <div class="code-summary">
    <strong>Feature Engineering:</strong>
    <ul>
      <li><strong>Fossil Fuels Deficit:</strong> Consumption minus production for fossil fuels.</li>
      <li><strong>Nuclear Power Deficit:</strong> Consumption minus production for nuclear power.</li>
      <li><strong>Renewable Energy Deficit:</strong> Consumption minus production for renewable energy.</li>
      <li><strong>Primary Energy Deficit:</strong> Total energy consumption minus production.</li>
    </ul>
  </div>
</body>
</html>


```python
energy_annual['Fossil Fuels Deficit'] = energy_annual['Total Fossil Fuels Consumption'] - energy_annual['Total Fossil Fuels Production']
energy_annual['Nuclear Power Deficit'] = energy_annual['Nuclear Electric Power Consumption'] - energy_annual['Nuclear Electric Power Production']
energy_annual['Renewable Energy Deficit'] = energy_annual['Total Renewable Energy Consumption'] - energy_annual['Total Renewable Energy Production']
energy_annual['Primary Energy Deficit'] = energy_annual['Total Primary Energy Consumption'] - energy_annual['Total Primary Energy Production']
```

```python
energy_annual.head()
```

<div style="font-family: Arial, sans-serif; line-height: 1.6; color: #333;">
  <p style="font-size: 18px;">In this code, we calculate the correlation matrix of the annual energy consumption data using the <code style="color: #D14F4F; background-color: #F9F2F4;">corr()</code> method. This matrix shows how different energy consumption metrics are related to each other.</p>

  <p style="font-size: 18px;">We then visualize the correlation matrix as a heatmap using the <code style="color: #D14F4F; background-color: #F9F2F4;">seaborn.heatmap()</code> function. The heatmap is displayed with annotations (values) on each cell, using the <code style="color: #D14F4F; background-color: #F9F2F4;">coolwarm</code> color map. The color range is set between -1 and 1, where -1 represents a perfect negative correlation, 1 represents a perfect positive correlation, and 0 represents no correlation.</p>
</div>


```python
corr_matrix = energy_annual.corr()

# Visualize the correlation matrix
plt.figure(figsize=(25, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Energy Consumption Metrics')
plt.show()
```

<div style="font-family: Arial, sans-serif; line-height: 1.6; color: #333;">
  <p style="font-size: 18px;">In this plot, we are analyzing the annual trends in energy consumption across various sectors, including <strong>Primary Energy Consumption</strong>, <strong>Fossil Fuels Consumption</strong>, <strong>Nuclear Electric Power Consumption</strong>, and <strong>Renewable Energy Consumption</strong>. Each line represents a different sector, with the x-axis showing the years and the y-axis representing the amount of energy consumed in each sector.</p>

  <p style="font-size: 18px;">By visualizing these trends, we can expect to learn the following insights:</p>
  <ul style="font-size: 18px;">
    <li><strong>Growth or Decline in Consumption:</strong> We can observe whether energy consumption in each sector has been increasing or decreasing over the years.</li>
    <li><strong>Sector Comparisons:</strong> The plot allows us to compare how different energy sectors (fossil fuels, nuclear, renewable, etc.) are performing relative to each other in terms of consumption.</li>
    <li><strong>Shifts in Energy Mix:</strong> If renewable energy consumption shows an upward trend while fossil fuels consumption declines, it may indicate a shift toward cleaner energy sources.</li>
    <li><strong>Impact of External Factors:</strong> Any sharp increases or decreases in consumption could correlate with external factors like policy changes, technological advancements, or economic shifts.</li>
  </ul>

  <p style="font-size: 18px;">This plot helps in understanding the evolution of energy consumption patterns and the ongoing transition towards sustainable energy sources.</p>
</div>


```python
plt.figure(figsize=(12, 6))
sns.lineplot(x="Annual Total", y="Total Primary Energy Consumption", data=energy_annual, label="Total Primary Energy Consumption")
sns.lineplot(x="Annual Total", y="Total Fossil Fuels Consumption", data=energy_annual, label="Total Fossil Fuels Consumption")
sns.lineplot(x="Annual Total", y="Nuclear Electric Power Consumption", data=energy_annual, label="Nuclear Electric Power Consumption")
sns.lineplot(x="Annual Total", y="Total Renewable Energy Consumption", data=energy_annual, label="Total Renewable Energy Consumption")

# Add informative titles and labels
plt.title('Annual Trends in Energy Consumption by Sector')
plt.xlabel('Year')
plt.ylabel('Energy Consumption (Units)')
plt.legend(title='Sector')

# Display the plot
plt.show()
```
